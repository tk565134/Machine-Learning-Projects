```python
import pandas as pd
```

```python
df = pd.read_csv('drive/My Drive/House Price Prediction/train.csv')
df.head()
```

```python
df.drop(columns='Id',inplace=True)
```

```python
df.isnull().sum()
```

```python
df.fillna(0)
df.shape
```

```python
ob = df.dtypes
ob = ob.replace(['int64','object','float64'],['numerical','category','numerical'])
x= pd.Series.to_dict(ob)
x
```

```python
del x['SalePrice']
```

```python
df1 = pd.DataFrame()
df1['name'] = x.keys()
df1['type'] = x.values()
df1.head()
```

```python
#df2.drop(columns='index',inplace=True)
x = df1.T.to_dict()
featurelist = []
for i in range(79):
    featurelist.append(x[i])
featurelist
```

```python
!pip install ludwig
```

```python
from ludwig.api import LudwigModel

model_definition = {
    'input_features':featurelist,
    'output_features': [
        {'name': 'SalePrice', 'type': 'numerical'}
    ]
}

print('creating model')
model = LudwigModel(model_definition)
print('training model')
train_stats = model.train(data_df=df)
model.close()
```

```python
model = LudwigModel.load('results/api_experiment_run_0/model/')
```

```python
type(model)
```

```python
test_data = pd.read_csv('drive/My Drive/House Price Prediction/test.csv')
#test_data.drop(columns='Id',inplace=True)
test_data.head()
```

```python
pred = model.predict(test_data)
type(pred)
```

```python
pred['Id'] = test_data.Id
pred.head()
```

```python
pred.rename(columns={'SalePrice_predictions':'SalePrice'},inplace=True)
pred.head()
```

```python
pred=pred[pred.columns[::-1]]
pred.head()
```
